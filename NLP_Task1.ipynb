# Task 4: Exploring Webtext corpus (Homework)

```
from nltk.corpus import webtext
nltk.download('webtext')
webtext.words(fileids = 'wine.txt')[:20]
```

# Task 5: Frequency Distribution

```
text = webtext.words(fileids = 'wine.txt')

fd = nltk.FreqDist(text)
fd
```

#Task 6: Conditional Frequency Distribution

```
from nltk.probability import ConditionalFreqDist

cfd = ConditionalFreqDist((len(word), word) for word in text)
```

```
cfd[5] #words with five character length
```

#Task 7: Freq distribution and Conditional Freq distribution for the innaugural speech of a US President (Homework)

```
meme_ref = '2017-Trump.txt'

some_speech = inaugural.words(fileids=meme_ref)

speech_length = len([word for word in some_speech if word not in "',./-?!"])
print(speech_length)

frequency_dist = nltk.FreqDist([word for word in some_speech if word not in "',./-?!"])
frequency_dist.most_common(50)
```

```
Obama = '2013-Obama.txt'

Obama_speech = inaugural.words(fileids=Obama)
Obama_speech_length = len([word for word in Obama_speech if word not in "',./-?!"])
print(Obama_speech_length)

Obama_fd = nltk.FreqDist([word for word in Obama_speech if word not in "',./-?!"])
Obama_fd.most_common(50)
```

Trump has a relatively shorter speech than Obama. Trump most commonly used **and** which denotes he likes conveying a lot of ideas, in one sentence maybe compared to obama, whose use of **the** hints more specifics to be pointed out instead of referring by a or an. Also use of words like 'America' or 'American' by Trump is more, denoting his general reference to the country more in speeches.
